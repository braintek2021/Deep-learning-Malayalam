```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import cv2
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path =r"C:\My_directory\knowledge\detection\trial/predict/"
images=os.listdir(path)
```

```python
def class_pred(path):
    for i in images:
        prediction=[]
        pred=model.predict(path+i)
        for res in pred:
            class_id=res.boxes.cls
            class_id = class_id.numpy()
            for item in class_id:
                class_name = coco_classes[item]
                prediction.append(class_name)
        return prediction
```

```python
def conf_pred(path):
    for i in images:
        confidence=[]
        pred=model.predict(path+i)
        for res in pred:
            confidence_score = res.boxes.conf
            confidence_score =confidence_score.numpy()
            print(confidence_score)
            for item in confidence_score:
                confidence.append(item) 
        return confidence
```

```python
class_pred(path)
```

```python
conf_pred(path)
```

```python
def image_plot(path):
    for i in images:
        img = cv2.imread(path + i)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        pred = model.predict(path + i)
        
        for res in pred:
            confidence_score = res.boxes.conf
            confidence_score = confidence_score.numpy()
            class_id = res.boxes.cls
            class_id = class_id.numpy()
            xyxy_boxes = res.boxes.xyxy.numpy()  # Get bounding box coordinates

            for idx, confidence, bbox in zip(class_id, confidence_score, xyxy_boxes):
                class_name = coco_classes[idx]
                x1, y1, x2, y2 = bbox

                x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                img_crop = img[y1:y2, x1:x2]
                
                cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
                print(class_name,confidence)
                cv2.putText(img, f"{class_name}: {confidence:.2f}", (x1+15, y1+15),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.3, (255, 255, 0), 1)

                plt.imshow(img_crop)
                plt.show()
```

```python
image_plot(path)
```
