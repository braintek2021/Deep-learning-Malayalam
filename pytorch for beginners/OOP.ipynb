# The __init__ method 

The __init__ method is similar to constructors in C++ and Java. It is run as soon as an object of a class is instantiated. The method is useful to do any initialization you want to do with your object. 

o understand the meaning of classes we have to understand the built-in __init__() function.

All classes have a function called __init__(), which is always executed when the class is being initiated.

Use the __init__() function to assign values to object properties, or other operations that are necessary to do when the object is being created:

```python
class car():
     
    # init method or constructor
    def __init__(self, model, color,V,t):
        self.model = model
        self.color = color
        self.V= V
        self.t= t
  
         
    def car_details(self):
        print("Model is", self.model )
        print("color is", self.color )
        
    def accln(self):
        a=self.V/self.t
        print("speed is", a )
   
        
         
# both objects have different self which
# contain their attributes
audi = car("audi a4", "blue",100,2)
ferrari = car("ferrari 488", "green",200,2)
 
audi.car_details() # same output as car.car_details(audi)
audi.accln()

 
```

```python
ferrari.car_details()  # same output as car.car_details(ferrari) 
ferrari.accln()
```

```python
car.car_details(audi)
car.accln(audi)
car.car_details(ferrari)
car.accln(ferrari)
```

# Python super() Function

- The super() function is used to give access to methods and properties of a parent or sibling class.

- The super() function returns an object that represents the parent class

```python
class Parent:
  def __init__(self, txt):
    self.message = txt

  def printmessage(self):
    print(self.message)

class Child(Parent):
  def __init__(self, txt):
    super().__init__(txt)

x = Child("Hello, and welcome!")

x.printmessage()
```
