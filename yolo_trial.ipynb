```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import cv2
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path =r"C:\My_directory\knowledge\detection\trial/predict/"
images=os.listdir(path)
```

```python
def class_pred(path):
    for i in images:
        prediction=[]
        pred=model.predict(path+i)
        for res in pred:
            class_id=res.boxes.cls
            class_id = class_id.numpy()
            for item in class_id:
                print(item)
                class_name = coco_classes[item]
                #conf=confidence_score[item]
                prediction.append(class_name)
            #confidence.append(conf) 
        return prediction
```

```python
def conf_pred(path):
    for i in images:
        confidence=[]
        pred=model.predict(path+i)
        for res in pred:
            confidence_score = res.boxes.conf
            confidence_score =confidence_score.numpy()
            print(confidence_score)
            for item in confidence_score:
                confidence.append(item) 
        return confidence
```

```python
class_pred(path)
```

```python
conf_pred(path)
```

```python
def image_plot(path):
    for i in images:
        img=cv2.imread(path+i)
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        pred=model.predict(path+i)
        for res in pred:
            cord=res.boxes.xyxy.numpy()
            for j in cord:
                x1,y1,x2,y2=j
                x1,y1,x2,y2=int(x1),int(y1),int(x2),int(y2)
                img_crop=img[y1:y2,x1:x2]
                clss,prob=class_pred(path),conf_pred(path)
                cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)
                cv2.putText(img,f'{clss} {str(prob)[:4]}',(x1-10,y1-10),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
            #plt.imshow(img)
                plt.imshow(img_crop)
                plt.show()
```

```python
image_plot(path)
```
